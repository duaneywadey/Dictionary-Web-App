```python
reg = linear_model.LinearRegression()
# reg.fit(df[['Day']], df.Prices)
x = np.array(df[['Day']])
y = np.array(df[['Price']])

reg.fit(x.reshape(-1, 1), y)
print(reg.predict([[22]]))




```

```python
plt.plot(df.Day, df.Price)
plt.show()
```
